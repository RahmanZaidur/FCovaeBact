```python
"""
Read fasta file into Dataframe
"""

import pandas as pd
from Bio import SeqIO

# Read the fasta file
records = list(SeqIO.parse("FCovae_ncbi1.faa", "fasta"))

# Prepare the lists to store the data
descriptions = []
sequences = []

# Loop over the records and save the id and sequence data
for seq_record in records:
    descriptions.append(seq_record.id)
    sequences.append(str(seq_record.seq))

# Create a dataframe
df1 = pd.DataFrame({"description": descriptions, "sequence": sequences})

# Display the dataframe
df1
```

```python
import pandas as pd

# List to store all the dataframes
dataframes = []

# Loop through the range of dataframe names
for i in range(1, 23):
    # Get the dataframe name dynamically using f-string
    df_name = f"df{i}"
    # Assuming the dataframes are already defined with the respective names
    # Add the dataframe to the list
    dataframes.append(locals()[df_name])

# Concatenate all the dataframes in the list
fcovae_df = pd.concat(dataframes)

# Reset the index of the new dataframe
fcovae_df = fcovae_df.reset_index(drop=True)
fcovae_df
```

```python
# Add the ID column with values in the desired format
fcovae_df.insert(0, "id", [f"seq{i}" for i in range(1, len(fcovae_df) + 1)])

# Print the updated dataframe
fcovae_df
```

```python
fcovae_df.to_csv("fcovae_df_description.csv", index=False)
```

```python
# Drop the description column
fcovae_df = fcovae_df.drop("description", axis=1)
fcovae_df
```
